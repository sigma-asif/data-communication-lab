```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
Ac = Am = A = 1     # Amplitude of carrier and message signals
fm = 1              # Message signal frequency (Hz)
fc = 10              # Carrier signal frequency (Hz)
sr = 100            # Sampling rate
t = np.linspace(0, 4, 4 * sr, endpoint=False)  # Time vector for 4 seconds

# Carrier Signal
def carrier_signal():
    return A * np.sin(2 * np.pi * fc * t)

# Message Signal
def msg():
    return A * np.sin(2 * np.pi * fm * t)

# Amplitude Modulation
def AM():
    return (Ac + Am * np.sin(2 * np.pi * fm * t)) * np.sin(2 * np.pi * fc * t)

# Frequency Modulation
def FM():
    kf = 20
    integral_msg = -Am / (2 * np.pi * fm) * np.cos(2 * np.pi * fm * t)
    st = Ac * np.sin(2 * np.pi * fc * t + kf * integral_msg)
    return st

# Phase Modulation
def PM():
    kp = np.pi
    st = Ac * np.sin(2 * np.pi * fc * t + kp * Am * np.sin(2 * np.pi * fm * t))
    return st


def main():
    # Message Signal
    signal = msg()
    plt.plot(t, signal)
    plt.title("Message Signal (1 Hz Sine Wave)")
    plt.xlabel("Time [s]")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.show()

    # Carrier Signal
    signal = carrier_signal()
    plt.plot(t, signal)
    plt.title("Carrier Signal (4 Hz Sine Wave)")
    plt.xlabel("Time [s]")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.show()

    # Amplitude Modulated Signal
    signal = AM()
    plt.plot(t, signal)
    plt.title("Amplitude Modulation (AM)")
    plt.xlabel("Time [s]")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.show()

    # Frequency Modulated Signal
    signal = FM()
    plt.plot(t, signal)
    plt.title("Frequency Modulation (FM)")
    plt.xlabel("Time [s]")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.show()

    # Phase Modulated Signal
    signal = PM()
    plt.plot(t, signal)
    plt.title("Phase Modulation (PM)")
    plt.xlabel("Time [s]")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.show()

# Run the program
main()
```
