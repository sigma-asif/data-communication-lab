```python
import numpy as np
import matplotlib.pyplot as plt

f = 1
T = 1/f
t = np.linspace(0,T,int(T*100),endpoint = False)

num_samples = 16
num_levels = 16

def analog_signal():
    return np.sin(2*np.pi*f*t)

def sampled_signal():

    ts = np.linspace(0,T,num_samples)
    ss = np.sin(2*np.pi*f*ts)

    return ts,ss

def quantized_signal():
    ts,ss = sampled_signal()
    normalisation = (ss-min(ss))/(max(ss)-min(ss))
    return ts,np.round(normalisation*num_levels).astype(int)

def binary_representation():
    ts,qs = quantized_signal()

    return [format(point,'04b') for point in qs]

def main():

    #analog signal
    y = analog_signal()
    plt.plot(t,y)
    plt.title("analog signal")
    plt.grid(True)
    plt.show()

    #sampled signal
    ts,ss = sampled_signal()
    plt.plot(ts,ss,'o-',color = 'red')
    plt.title("Sampled Signal")
    plt.grid(True)
    plt.show()

    #quantized signal
    ts,qs = quantized_signal()
    plt.plot(ts,qs,'o-',color = 'red')
    plt.title("Quantized Signal")
    plt.grid(True)
    plt.show()

    #quantized signal
    ts,qs = quantized_signal()
    plt.stem(ts,qs)
    plt.title("Quantized sampled Signal")
    plt.grid(True)
    plt.show()

    #printing
    binary_values = binary_representation()
    for value, binary in zip(qs, binary_values):
        print(f"value: {value}, binary: {binary}")

main()
```
